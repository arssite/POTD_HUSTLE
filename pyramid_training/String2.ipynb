```python
def minimize_frames(T, cases):
    results = []
    for case in cases:
        N, frames = case
        if N == 1:
            results.append(1)
            continue

        # Initialize the list of unique consecutive frames
        minimized_frames = [frames[0]]

        for i in range(1, N):
            if frames[i] != frames[i - 1]:
                minimized_frames.append(frames[i])

        results.append(len(minimized_frames))

    return results

# Read input
T = int(input().strip())
cases = []

for _ in range(T):
    N = int(input().strip())
    frames = list(map(int, input().strip().split()))
    cases.append((N, frames))

# Get results
results = minimize_frames(T, cases)

# Output results
for result in results:
    print(result)
```
